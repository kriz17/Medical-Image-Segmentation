```python
import numpy as np 
import pandas as pd
import os
import shutil
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
```

# Creating custom folders for data for easier access

```python
#SOURCE PATHS
TRAIN_PATH = 'data/stage1_train/'
VAL_PATH = 'data/stage1_test/'
TEST_PATH = 'data/stage2_test_final/'

train_ids = next(os.walk(TRAIN_PATH))[1]
val_ids = next(os.walk(VAL_PATH))[1]
test_ids = next(os.walk(TEST_PATH))[1]

print('number of train instances:',len(train_ids))
print('number of validation instances:',len(val_ids))
print('number of test instances:',len(test_ids))
```

```python
#SAVING PATHS OF IMAGES AND MASKS
train_x_paths = []
for id in train_ids:
  train_x_paths.append(TRAIN_PATH + id + '/images/' + id + '.png')
train_y_paths = []
for id in train_ids:
  train_y_paths.append(TRAIN_PATH + id + '/merged/' + id + '.png')
val_x_paths = []
for id in val_ids:
  val_x_paths.append(VAL_PATH + id + '/images/' + id + '.png')
val_y_paths = []
for id in val_ids:
  val_y_paths.append(VAL_PATH + id + '/merged/' + id + '.png')
test_x_paths = []
for id in test_ids:
  test_x_paths.append(TEST_PATH + id + '/images/' + id + '.png')

print('train x and y:', len(train_x_paths), len(train_y_paths))
print('val x and y:', len(val_x_paths), len(val_y_paths))
print('test x:', len(test_x_paths))
```

```python
#DESTINATION PATHS
dest_train_set = 'data_custom/train/'
dest_val_set = 'data_custom/val/'
dest_test_set = 'data_custom/test/'
```

```python
#COPY IMAGES FROM SOURCE TO DESTINATION
flag = False

if flag:
  for f in tqdm(train_x_paths):
    shutil.copy(f, dest_train_set + 'images')
  for f in tqdm(train_y_paths):
    shutil.copy(f, dest_train_set + 'masks')
  for f in tqdm(val_x_paths):
    shutil.copy(f, dest_val_set + 'images')
  for f in tqdm(val_y_paths):
    shutil.copy(f, dest_val_set + 'masks')
  for f in tqdm(test_x_paths):
    shutil.copy(f, dest_test_set + 'images')
```
